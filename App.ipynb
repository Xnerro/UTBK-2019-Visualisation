```python
df_universitas = pd.merge(df_fakultas, df_univ[['id_university', 'university_name']], left_on = 'id_university', right_on = 'id_university')
df_universitas.drop(['Unnamed: 0'], axis = 1, inplace = True)

df_universitas.head()
```

```python
df_universitas['major_univ_name'] = df_universitas['major_name'] + ' - ' + df_universitas['university_name']

df_universitas.head()
```

```python
df_ips['avg_score'] =  df_ips.iloc[:, 6:15].mean(axis=1)

df_ips.head()
```

```python
df_ipa['avg_score'] =  df_ipa.iloc[:, 6:14].mean(axis=1)

df_ipa.head()
```

```python
df_ips = pd.merge(df_ips, df_universitas[['id_major', 'id_university', 'major_name', 'university_name', 'major_univ_name']], left_on = ['id_first_major', 'id_first_university'], right_on = ['id_major', 'id_university']).drop(['id_major','id_university' , 'id_first_major', 'id_first_university'], axis=1)
```

```python
df_ips.rename(columns={'major_univ_name' : 'first_choice'}, inplace=True)
df_ips['avg_score_first'] = df_ips.groupby('first_choice')['avg_score'].transform('mean')
```

```python
df_ips = pd.merge(df_ips, df_universitas[['id_major', 'id_university', 'major_name', 'university_name', 'major_univ_name']], left_on = ['id_second_major', 'id_second_university'], right_on = ['id_major', 'id_university']).drop(['id_major','id_university' , 'id_second_major', 'id_second_university'], axis=1)
```

```python
df_ips.rename(columns={'major_univ_name' : 'second_choice'}, inplace=True)
df_ips['avg_score_secod'] = df_ips.groupby('second_choice')['avg_score'].transform('mean')
```

```python
df_ips.drop(['major_name_x','university_name_x' ,'major_name_y', 'university_name_y'], axis=1)
```

```python
df_ipa = pd.merge(df_ipa, df_universitas[['id_major', 'id_university', 'major_name', 'university_name', 'major_univ_name']], left_on = ['id_first_major', 'id_first_university'], right_on = ['id_major', 'id_university']).drop(['id_major','id_university' , 'id_first_major', 'id_first_university'], axis=1)
```

```python
df_ipa.rename(columns={'major_univ_name' : 'first_choice'}, inplace=True)
df_ipa['avg_score_first'] = df_ips.groupby('first_choice')['avg_score'].transform('mean')
```

```python
df_ipa = pd.merge(df_ipa, df_universitas[['id_major', 'id_university', 'major_name', 'university_name', 'major_univ_name']], left_on = ['id_second_major', 'id_second_university'], right_on = ['id_major', 'id_university']).drop(['id_major','id_university' , 'id_second_major', 'id_second_university'], axis=1)
```

```python
df_ipa.rename(columns={'major_univ_name' : 'second_choice'}, inplace=True)
df_ipa['avg_score_second'] = df_ips.groupby('second_choice')['avg_score'].transform('mean')
```

```python
df_ipa.drop(['major_name_x','university_name_x' ,'major_name_y', 'university_name_y'], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
fig.tight_layout(pad=5)

sns.distplot(ax = ax, a = df_universitas['capacity'].dropna(), label = 'skewness : %.2f'%(df_universitas['capacity'].skew()))
ax.set_title('capacity', fontsize=18)
ax.legend(loc = 'best')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,6))

sizes = [count for count in df_universitas['type'].value_counts()]
labels = list(df_universitas['type'].value_counts().index + ' major')
ax.pie (x = sizes, labels = labels, autopct = '%1.1f%%', textprops = {'fontsize':14})

plt.show
```

```python
fig, ax = plt.subplots(2,1 ,figsize = (14,8))
fig.tight_layout(pad = 6)
max_width = 13

index = df_universitas['major_name'].fillna('NaN').value_counts().index[0:10]
count = df_universitas['major_name'].fillna('NaN').value_counts()[0:10]
sns.barplot(ax = ax[0], x = index, y = count, order = index)
ax[0].set_title('Top 10 Most Majors are Provided by Universities', fontsize = 18)
ax[0].set_xticklabels((textwrap.fill(x.get_text(), max_width) for x in ax[0].get_xticklabels()), fontsize = 10)
for index,count in enumerate(count):
    ax[0].text(x=index-0.1 , y =count+0 , s=f"{count}" , fontdict=dict(fontsize=10))
ax[0].set_ylabel('No. of Universities', fontsize = 14)
ax[0].set_ylim(40, 65)

index = df_universitas['university_name'].fillna('NaN').value_counts().index[0:10]
count = df_universitas['university_name'].fillna('NaN').value_counts()[0:10]
sns.barplot(ax = ax[1], x = index, y = count, order = index)
ax[1].set_title('Top 10 Universitas Berdasakarkan Fakultas', fontsize = 18)
ax[1].set_xticklabels((textwrap.fill(x.get_text(), max_width) for x in ax[1].get_xticklabels()), fontsize = 10)
for index , count in  enumerate(count):
  ax[1].text(x=index-1.0 , y = count+0, s=f"{count}", fontdict = dict(fontsize = 10))
ax[1].set_ylabel(' No. Fakultas', fontsize = 14)
ax[1].set_ylim(60, 85)

plt.show()
```

```python
num_df_ips = ['score_eko', 'score_geo', 'score_kmb', 'score_kpu', 'score_kua', 'score_mat', 'score_ppu', 'score_sej', 'score_sos', 'avg_score']

fig, ax = plt.subplots(5,2, figsize = (14,14))
fig.tight_layout(pad = 5)

for ax, n in zip(ax.flatten(), num_df_ips):
  sns.distplot(ax = ax, a = df_ips[n].dropna(), label = "Skewness : %.2f"%(df_ips[n].skew()))
  ax.set_title(n, fontsize = 18)
  ax.legend(loc = 'best')

plt.show()
```

```python
cormat = df_ips[num_df_ips].corr()

fig, ax = plt.subplots(figsize = (12,8))
sns.heatmap(ax = ax, data = cormat, annot = True)
ax.set_yticklabels(cormat, rotation = 0)

plt.show()
```
